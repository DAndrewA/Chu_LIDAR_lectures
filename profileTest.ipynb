# Test for the atmospheric profiles

I'll define a set of points for heights up to 70km that I will use to test the atmospheric profiles defined in this folder.

```python
import numpy as np
import matplotlib.pyplot as plt

# importing the profiles
import profiles

z = np.arange(0,70e3,100)
```

```python
# code to run and plot the US_standard_atmosphere_1976 profile

T,p,rho = profiles.US_standard_atmosphere_1976(z,k='quadratic')

fig = plt.figure()
ax = fig.add_subplot(1,3,1)
ax.plot(T,z)
ax = fig.add_subplot(1,3,2)
ax.plot(p,z)
ax = fig.add_subplot(1,3,3)
ax.plot(rho,z)

plt.show()
```

```python
# testing the Isothermal_hydrostatic profile
T,p,rho = profiles.Isothermal_hydrostatic(z)

fig = plt.figure()
ax = fig.add_subplot(1,3,1)
ax.plot(T,z)
ax = fig.add_subplot(1,3,2)
ax.plot(p,z)
ax = fig.add_subplot(1,3,3)
ax.plot(rho,z)

plt.show()
```
